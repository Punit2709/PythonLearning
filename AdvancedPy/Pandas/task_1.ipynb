```python
import pandas as pd
```

### Qn 1 
- Reading CSV File

```python
df = pd.read_csv('../Pandas/sample_data/sales_data.csv')
df.head()
```

### Q2
- Find all orders where order_amt is greater than 500.

```python
df_gtr_500 = df[df['order_amt'] > 500]
df_gtr_500
```

### Q3
- Sort the DataFrame by profit_amt 

```python
df_sorted = df.sort_values(by="profit_amt", ascending=False)
df_sorted.head()
```

### Qn 4
- Find the total order_amt for each province.

```python
sales_by_province = df.groupby("province")["order_amt"].sum().reset_index()
sales_by_province
```

### Qn 5
- Get a list of unique product_type values in the dataset.

```python
df_unique_product = df['product_type'].unique()
df_unique_product
```

### Qn 6
- Count and display the number of missing values in each column.

```python
df_missing = df.isna().sum()
df_missing
```

### Qn 7
- Filter out all orders placed in the year 2022.

```python
df_2022 =df[pd.to_datetime(df['report_date']).dt.year == 2022]
df_2022
```

### Qn 8
- Create a new column profit_margin calculated as (profit_amt / order_amt) * 100.

```python
df['profit_mgn'] = df['profit_amt'] / df['order_amt'] * 100
df
```

### Qn 9
- Find the most common shipping_type used.

```python
df_cmn_shipps = df['shipping_type'].value_counts()
df_cmn_shipps
```

### Qn 10
- Create a pivot table showing the total order_amt for each province and city.

```python
pvt_df = df.pivot_table(index=['province', 'city'], values='order_amt', aggfunc='sum')
pvt_df
```
