### Qn 5
- Fetch all job titles that contain the word "Engineer".

```python
import re
ptrn = r"\b(Engineer|engineer)\b"
df_eng = ecom_df['Job']
list = []
for job in df_eng:
    if re.search(ptrn, job):
        list.append(job)
    
list
```

### Qn 9 
- What are the top 5 most popular email providers (e.g., Gmail, Yahoo)?

```python
ecom_df['email_domain'] = ecom_df['Email'].str.split('@').str[1]
ecom_df
```

```python
email_df = ecom_df.groupby('email_domain')['Email'].count().reset_index()
email_df.rename(columns={'Email': 'count'}, inplace=True)
email_df = email_df.sort_values(by='count', ascending=False)
email_df.head(5)
```

### Qn 10

### Qn 11
- Find the most common job title in the dataset.

```python
job_df = ecom_df.groupby('Job')['Address'].count().reset_index()
job_df.rename(columns={'Address': 'count'}, inplace=True)
job_df = job_df.sort_values(by='count', ascending=False)
job_df
```
