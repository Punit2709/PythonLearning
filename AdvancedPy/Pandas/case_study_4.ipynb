```python
import pandas as pd
```

```python
movie_df = pd.read_csv('../Pandas/sample_data/IMDB_Movies_DATA.csv')
movie_df
```

### Qn 1
- Drop all missing values from the dataset.

```python
drop_df = movie_df.dropna()
drop_df
```

### Qn 2
- Check for duplicate data in the dataset.

```python
duplecated_df = drop_df.duplicated()
duplecated_df
```

### Qn 3
- Get overall statistics about the dataset.

```python
stats = drop_df.describe()
stats
```

### Qn 4
- Display the title of movies having a runtime of 180 minutes or more.

```python
long_movies = drop_df[drop_df['Runtime (Minutes)'] > 180]
long_movies[['Title', 'Runtime (Minutes)']]
```

### Qn 5
- Fnd the year with the highest average voting. (Group by "Year" and get the mean of "Votes".)

```python
drop_df['Votes'] = drop_df['Votes'].fillna(0)  # or dropna()
drop_df['mean_votes'] = drop_df.groupby('Year')['Votes'].agg('mean')
drop_df
# high_vote = drop_df.sort_values(by='mean_votes',ascending=False)
# high_vote.head(1)
```

### Qn 6
- Find the year with the highest average revenue. (Group by "Year" and get the mean of "Revenue”)

```python
drop_df['Revenue (Millions)'] = drop_df['Revenue (Millions)'].fillna(0)  # or dropna()
drop_df['mean_rev'] = drop_df.groupby('Year').agg('mean')
drop_df
```

### Qn 7
- Find the average rating for each director.

```python
df_avg_dir = drop_df.groupby('Director')['Rating'].agg('mean')
df_avg_dir.sort_values(ascending=False)
# df_avg_dir.info()
```

### Qn 8
- Display the top 10 longest movies by title and runtime.

```python
df_long_movie = drop_df.sort_values(by='Runtime (Minutes)', ascending=False).head(10)
df_long_movie[['Title', 'Runtime (Minutes)']]
```

### Qn 9

### Qn 10

```python
df_high_rev = drop_df.sort_values('Revenue (Millions)', ascending=False).head(1)
df_high_rev
```

### Qn 11
