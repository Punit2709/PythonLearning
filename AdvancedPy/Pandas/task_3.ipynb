### Qn 1
- Load Data Into Dataframe

```python
df = pd.read_csv('../Pandas/sample_data/adult_salaries.csv')
df.head()
```

### Qn 2

```python
print(df.shape)
```

### Qn 3
- check for null values

```python
df_null = df.isna()
df_null
```

### Qn 4
- replace '?' by NaN

```python
df_replica = df.replace('?', pd.NA)
df_replica
```

### Qn 5
- Drop Row With Missing Value

```python
df_droped = df_replica.dropna()
df_droped
```

### Qn 6
- Dataset Statisticks

```python
df_stat = df_droped.describe()
df_stat
```

### Qn 7
- Drop Columns education-num, capital-gain, and capital-loss.

```python
df_droped_col = df_droped.drop(columns=['educational-num', 'capital-gain', 'capital-loss'])
df_droped_col
```

### Qn 8
- Find Pesron having age btw 17 and 47 (inclusive)

```python
df_adult = df[(df['age'] > 17) & (df['age'] <= 47)]
df_adult.head(50)              
```

### Qn 9
- Person With Bachelor or Master Degree

```python
df_degree = df[(df['education'] == 'Bachelors') | (df['education'] == 'Masters')]
df_degree.head(100)
```

### Qn 10

```python
df['income'] = df['income'].replace({'<=50K': 0, '>50K': 1})
df
```
